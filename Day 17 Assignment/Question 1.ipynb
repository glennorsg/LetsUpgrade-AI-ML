```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot
import seaborn as sns
cars = pd.read_csv("mtcars.csv")
cars
```

```python
print(cars.describe())
```

```python
cars.isnull().any()
```

```python
sns.boxplot(cars.hp)
```

```python
cars_df = cars.drop('model', axis = 1)

z = np.round(np.abs(stats.zscore(cars_df)), decimals = 2)
print(z)
```

```python
print(np.where(z > 3))
```

```python
z[30][10]
```

From the above analysis using zscore there is an outliar on the 30th record in the carb column
